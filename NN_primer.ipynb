A NN is a simplified representation of an aggregation of brain cells. The NODEs (circles) simulate neurons; the lines connecting them, hereafter referred to as EDGEs, correspond to synapses. Edges take the value of a node, multiply by a number we will call, a WEIGHT and transmit the output to the next node downstream.

```python
Embed=Image(filename="/disk01/home/ekarakoy/BANDSHIFTING/nnipynb_2.png",embed=True)
Embed
```

The nodes will aggregate the outputs from incoming edges, add them up(1) and apply them to an activation function (2):

$(1) z = x_1 + x_2 + ... x_n$

$(2) a = 1 \frac{1}{1+e^{-z}}$

Note that the activation function like sigmoid shown above (2) allows the NN to model COMPLEX NONLINEAR PATTERNS that simpler models might miss. In this example, I use a sigmoid activation function. I show an example below

```python
%matplotlib inline
def sigmoid(z):
    return 1/(1+np.exp(-z))
testInput=np.arange(-6,6,0.01)
plt.plot(testInput,sigmoid(testInput),lw=2)
plt.grid(1)
```

Now on to build the NN. Here we use the following indices:
* $i$: Data set index. This is just the row in our tables displayed above
* $j$: Input layer neuron index
* $k$: Hidden layer neuron index. If more than one, it'll be $k_1$, $k_2$,...,$k_n$
* $l$: Output layer neuron index

Given our input data (X) and output data (Y) data; the computations that take place

$$X_{ij} W^{(1)}_{jk} = Z^{(2)}_{ik}$$

where *i* is the row number, *j* is the column, i.e. band index, W holds the weights and k indexes the nodes in the 2nd (hidden) layer and Z^{(2)} is the output computed in the second (hidden layer)
The output at Z^{(2)} is fed to the sigmoid function such that  

$$ a^{(2)}_{ik} = f(Z^{(2)}_{ik})$$

Then we apply another weighing toward the final neuron, the sum of which will again be fed to another activation function, the output of which is the output of our output layer.

$$a^{(2)}_{ik} W^{(2)}_{kl} = Z^{(3)}_{il} $$
where, since we have only one output neuron, $l=1$
and
$$\hat{y} = f(Z^{(3)}_i)$$

A relatively simple implementation in python follows

#### Implementation in Python

```python
class Neural_Network(object):
    def __init__(self):
        # Define Hyperparameters so= called because they are fixed, defining the structure.
        # The structure remains fixed, i.e. NNs do not learn its geometry
        self.inputLayerSize=6
        self.outputLayerSize=1
        self.hiddenLayerSize=3
    
        # What NNs do learn are the parameters, specifically the weights on the edges (synapses)
        self.W1 = np.random.rand(self.inputLayerSize,self.hiddenLayerSize)
        self.W2 = np.random.rand(self.hiddenLayerSize,self.outputLayerSize)
    
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
    
    def forward(self,X):
        # This method takes care of moving the data through the network 
        # The data will be moved all in one chunk using matrix operations
        self.z2=np.dot(X,self.W1)
        self.a2=self.sigmoid(self.z2)
        self.z3=np.dot(self.a2,self.W2)
        yHat = self.sigmoid(self.z3)
        return yHat
```

To improve my model I need to quantify how wrong the predictions $\hat{y}$ are relative to $y$, using a cost function.
One way to do this is to compute the error for each prediction, square it and sum all the errors.
Thus $J=\frac{1}{2}\sum(y-\hat{y})^2$. The $\frac{1}{2}$ factor simplifies things further down the road.
Training the network means essentially, minimizing the cost function, $J$. $J$ here is a function of (1) the examples provided as training set, and the weights of the edges connecting the various nodes. We can't really control the data, so we attempt to improve the NN by adjusting the weights to minimize $J$.

Conceptually the idea would be simply to adjust all the weights by brute force until the best solution is found. One would expect a computer to make this pretty easy. This is without accounting for the *Curse of Dimensionality*. A description of the problem follows...

#### The curse of dimensionality

Say we have one weight, $W$ instead of the $6*3+3*1 = 21$ that our NN currently has and we are looking for a value for $W$ that will minimize the cost. Let's say we want to test 1000 values, this doesn't seem so bad, given that my computer can churn through it pretty quickly:

```python
import time
weightsToTry = np.linspace(-10,10,1000)
costs=np.zeros(1000)
startTime=time.clock()
for i in range(1000):
    NNchl.W1[0,0] = weightsToTry[i]
    cHat=NNchl.forward(nn_input_training_norm)
    costs[i] = 0.5*sum((nn_output_training_norm-cHat)**2)
endTime=time.clock()
timeElapsed = endTime-startTime
timeElapsed
```

Since we've tried a wide range of values for $W$, we can pick the one that minimizes our cost function:

```python
plt.plot(weightsToTry,costs)
plt.grid(1)
plt.ylabel('Cost')
plt.xlabel('Weight');
```

Looks like $W=-10$ is our winner. However, if we add a second weight, that corresponds to $1*10^6$ values

```python
weightsToTry = np.linspace(-10,10,1000)
costs=np.zeros((1000,1000))
startTime=time.clock()
for i in range(1000):
    for j in range(1000):
        NNchl.W1[0,0] = weightsToTry[i]
        NNchl.W1[0,1] = weightsToTry[j]
        cHat=NNchl.forward(nn_input_training_norm)
        costs[i,j] = 0.5*sum((nn_output_training_norm-cHat)**2)
endTime=time.clock()
timeElapsed = endTime-startTime
print timeElapsed
```

I'll just stop it as this would take way longer, and that's because the curse of dimensionality starts kicking in. As we add more and more weights, that corresponds to more and more combinations for the computer to try. In other words, sifting through all the possible combinations of weights for this NN, would amount to wait...

```python
0.91*(1000**(21-1))/(3600*24*365)
```

...years!! Nothing's that old, not even the universe. So the brute force method will not work. "Premature optimization is the root of all evil" (D. Knuth)

Let's try to add a little more information to what we already know. Let's look at the equations we have:
1. $z^{(2)} = XW^{(1)}$
2. $a^{(2)} = f(z^{(2)})$
3. $z^{(3)} = a^{(2)}W{(2)}$
4. $\hat y = f(z^{(3)})$
5. $J=\sum\frac{1}{2}(y-\hat y)^2$

This can be compacted into one bigger equation:
$$J=\sum \frac{1}{2} ( y -f(f(XW^{(1)})W^{(2)}))^2$$
What we can do then is examine the change in $J$ with respect to each weight $W$, which amounts to calculate partials.
If $\frac{\partial J}{\partial W} > 0$, the cost function is going up; otherwise, it's going down. THis enables a drastic speed up by restricting the search for appropriate weights in the direction that results in a cost decrease and stop where the decrease becomes insignificant. This is called **GRADIENT DESCENT**. This is especially impressive in higher dimensions.

The reason for using a parabola as a cost function is to take advantage of its generally convex shape to find a minimum. But what if the cost function is not convex, and local minima exist resulting in gradient descent being stuck in one of those local minimum?
One way to get around that is to use our data one at a time; what is known as stochastic gradient descent. But for now we will use batch gradient descent; i.e. all data are used at once, looking at $\sum\frac{\partial J}{\partial W}$. Back to our NN then...
